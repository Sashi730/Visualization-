```python
import matplotlib.pyplot as plt
```

## Shortcut for matplotlib Library

```python
'ro': oval with red color show.
'b-': line with blue color show.
'g--': Dotted line with green color show.
'y^': Triangle point with yellow color show.
plot(): To plot the graph.
xlabel(): To show the content in x-label
ylabel(): To show the content in y-label.
title(): To show the content in title bar.
grid(): To show the grid.
legend(): To show the legend.
show(): It used to show the graph.
linewidth: it show the the linewidth of the line.
```

## Draw a Graph using matplotlib framework

```python
plt.plot([1,4,6,8],'o--',linewidth=7,label='2^')
plt.xlabel("Indices")
plt.ylabel("Number")
plt.title("Graph")  # Everything that you give in the label of the plot that particular vslue show in your legend.
plt.grid()
plt.legend()
plt.show()
```

```python
l1=range(1,10)
l2=range(9)
plt.plot(l1,l2,'o--r')
plt.xlabel("Index")
plt.ylabel("Number")
plt.title("Graph")
plt.grid()
plt.show()
```

```python
plt.figure(1)
plt.subplot(2,1,1)
plt.plot(l1,'r--',linewidth=3)
plt.subplot(2,1,2)
plt.plot(l2,'y-')

plt.figure(2)
plt.plot([1,5,9,90],'g')

plt.figure(1)
plt.subplot(2,1,1)
plt.title("First Sub plot")
plt.show()
```

```python
plt.figure(1)
plt.subplot(2,2,1)
plt.plot(l2,'r--',linewidth=3)
plt.subplot(2,2,2)
plt.plot(l2,'yo')
plt.subplot(2,2,3)
plt.plot(l2,'g^')
plt.subplot(2,2,4)
plt.plot(l2,'b--')



plt.figure(1)
plt.subplot(2,2,1)
plt.title("First Sub plot")
plt.figure(1)
plt.subplot(2,2,2)
plt.title("Second Sub plot")
plt.figure(1)
plt.subplot(2,2,3)
plt.title("Third Sub plot")
plt.figure(1)
plt.subplot(2,2,4)
plt.title("Fourth Sub plot")
plt.suptitle("Whole Graph") # The title add in to the entire plot.
plt.show()
```

```python
import numpy as np
a1=np.array([1,2,9,10])
a2=np.array([8,3,5,16])
plt.plot([1,4,6],[5,8,19],'*:r',ms=19,mec='blue',mfc='cyan')
plt.plot(a1,a2,marker='o',markersize=20,mfc='k',mec='red') # if you use only marker point i.e 'o','*' then only point is shown in the graph.

plt.xlabel("Indices")
plt.ylabel("Number")
plt.title("Graph",loc='center') # Default value is center.
plt.grid(axis='both',color='k',ls='--',lw=1.3)
plt.show()
```

## Scatter Chart in matplotlib

```python

import numpy as np

x = np.array([5,7,8,7,2,17])
y = np.array([99,86,87,88,111,86])
sizes=np.array([100,40,70,80,30,90])# You can set color and size of each shape through an array.
plt.scatter(x, y,color='pink',s=sizes,alpha=0.9) # Alpha is used for the tranparency that ranges from 0-1.
plt.colorbar() # To plot The color bar.
plt.show()
```

## Barchart in matplotlib

```python
x=['A','B','C','D']
y=[10,20,30,40]
plt.figure(1)
plt.subplot(1,2,1)
plt.title("Vertical Bar")
plt.bar(x,y,color='pink',width=0.5,) # It create a vertical bar Chart
plt.subplot(1,2,2)
plt.barh(x,y,color='c',height=0.5,) # It create a vertical bar Chart
plt.title("Horizontal Bar")
plt.suptitle("Bar Chart")

plt.show()
```

## Create Pie chart

```python
y=[10,40,30,20]
mylabels=["Apple","Banana","Orange","Mango"]
mycolors=["red","black","green","cyan"]
myexplodes=[0.2,0,0,0]
plt.pie(y,labels=mylabels,colors=mycolors,shadow=True,startangle=90,explode=myexplodes) #Explode is used to represent how many distance the slice is present.
plt.legend(title="Fruits Items") # Add a legend title is the header of the legend.

```
