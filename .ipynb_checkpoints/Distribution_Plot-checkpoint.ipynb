# Distribution Graph

```python
# Create a random number.
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

p1=np.random.rand(10)
print(p1)

p2=np.random.randint(10,100,size=(10))
print(p2)

## Data Distribution.
p3=np.random.choice([1,3,4,5],p=[0.5,0.3,0.2,0.0],size=(2,3)) # Here p is the probability. and you have to set probability of each  element in an array
                                                             # The 1 has The highest probability and 0 has the probability.
print(p3)

# It return the random elements from the array.
p4=np.random.choice([10,20,30,40,40],size=(3))
print(p4)

sns.displot(np.random.rand(10))  # By default the rand is return a vertical  barchart.
plt.show()

sns.displot(np.random.randint(100,size=(20)))  # By default the rand is return a vertical barchart.
plt.show()

p5=np.random.normal(loc=1,scale=2,size=(13))
print(p5)

sns.displot(p5,kind="kde")  # By default the normal distribution given a bar or histogram.
plt.show()

p6=np.random.binomial(n=10,p=0.5,size=(5,6))  # kde means it show a Curve.
print(p6)

sns.displot(p6,kind="kde")  # By defult it show a bar like strure graph. Here could not pass 3D Input if you pass then it gives error.
plt.show()
```
